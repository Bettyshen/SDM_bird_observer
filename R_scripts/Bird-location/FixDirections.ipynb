## Read Data

```python
import pandas as pd
pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', None)  # or 199

data = pd.read_csv('Data/oregon2020-zerofilled.csv', encoding="cp1252")
print(len(data))
print(data.columns)
display(data.head(5))
```

```python
data['Common_Name'].value_counts()
```

```python
print(data['Direction'].unique())
print(data['Direction'].value_counts(dropna=False))
```

## Read Correct direction file

```python
corrections = pd.read_excel('Data/Direction_correct.xlsx')
corrections = corrections.drop(columns=['notes'])

data['Direction'] = data['Direction'].astype(str)
corrections['Direction'] = corrections['Direction'].astype(str)

print(corrections['Direction'].unique())
print(data['Direction'].unique())

display(corrections)
```

## Merge both datasets to update the correct direction

```python
result_df = pd.merge(data, corrections, on='Direction', how='left')
data['correct_direction'] = result_df['correct_direction'].combine_first(data['Direction'])
data.insert(3, 'correct_direction', data.pop('correct_direction'))
data['correct_direction'] = data['correct_direction'].str.upper()
data['correct_direction'] = data['correct_direction'].str.replace(' ', '')
```

```python
print(data['correct_direction'].unique())
print(data['correct_direction'].value_counts())
```

```python
# Define the values to drop
values_to_drop = ['SEE_NOTES', 'NEW', 'WW', 'NWN', 'SWW', 'CE', 'SS', 'EW', 'SN', '`N']  # Add any other values you want to drop

# Drop rows where correct_direction is in values_to_drop
data = data[~data['correct_direction'].isin(values_to_drop)]
print(len(data))
```
